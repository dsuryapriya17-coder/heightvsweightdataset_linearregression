```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/score.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10,10))
sns.pairplot(data,x_vars=['Hours'],y_vars=['Scores'],size=7,kind='scatter')
plt.xlabel=('Hours')
plt.ylabel=('Scores')
plt.show()
```

```python
x=data['Hours']
x.head()
```

```python
y=data['Scores']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_train_predict=lr.predict(x_train.values.reshape(-1,+1))
```

```python
from sklearn.metrics import mean_squared_error
mse_train=mean_squared_error(y_train,y_train_predict)
```

```python
from sklearn.metrics import r2_score
r2_train=r2_score(y_train,y_train_predict)
```

```python
print('Mse',mse_train)
print('accuracy',r2_train)
```

```python
c=[i for i in range(1,len(y_test)+1)]
y_predict=lr.predict(x_test.values.reshape(-1,+1))
plt.plot(c,y_test,color='r',linestyle='-')
plt.plot(c,y_predict,color='b',linestyle='-')
plt.x_label('Hours')
plt.y_label('Scores')
plt.title('Hours vs Scores')
plt.show()
```

```python
c=[i for i in range(1,len(y_test)+1)]
plt.plot(c,y_test-y_predict,color='green',linestyle='-')
plt.x_label('Hours')
plt.y_label('Scores')
plt.title('Error value')
plt.show()
```
